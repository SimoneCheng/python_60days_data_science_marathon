1.生成一個等差數列，首數為0，尾數為20，公差為1的數列。
2.呈上題，將以上數列取出偶數。
3.呈1題，將數列取出3的倍數。

```python
import numpy as np

#1.生成一個等差數列，首數為0，尾數為20，公差為1的數列。
a = np.arange(0,21,1)
a
```

```python
#2.呈上題，將以上數列取出偶數。
b = a[2:21:2]
b
```

```python
#3.呈1題，將數列取出3的倍數。
c = a[3:21:3]
c
```
