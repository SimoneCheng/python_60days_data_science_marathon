```python
#作業目標：
#敘述統計量計算
#運用自定義函數apply

#作業重點：
#了解敘述統計量並解釋
#運用apply時須注意自定義函數寫法

#題目：
#對以下成績資料做分析
import pandas as pd
score_df = pd.DataFrame([[1,56,66,70], 
              [2,90,45,34],
              [3,45,32,55],
              [4,70,77,89],
              [5,56,80,70],
              [6,60,54,55],
              [7,45,70,79],
              [8,34,77,76],
              [9,25,87,60],
              [10,88,40,43]],columns=['student_id','math_score','english_score','chinese_score'])
score_df = score_df.set_index('student_id')
score_df
```

```python
#6號學生(student_id=6)3科平均分數為何?
score_df.mean(axis=1)
#6號學生(student_id=6)3科平均分數為56.333333
```

```python
#6號學生3科平均分數是否有贏過班上一半的同學?
three_avr = score_df.mean(axis=1)
three_avr.median()
#3科平均分數的中位數是59.833333333333336
#6號學生的3科平均分數低於中位數，所以他的成績沒有贏過班上一半的同學
```

```python
#由於班上同學成績不好，所以學校統一加分，加分方式為開根號乘以十，請問6號同學3科成績分別是?
score_df.apply(lambda x : x**(0.5)*10)
#6號同學的math_score=77.459667；english_score=73.484692；chinese_score=74.161985
```

```python
#承上題，加分後各科班平均變多少?
new_score = score_df.apply(lambda x : x**(0.5)*10)
new_score.mean()
#加分後的全班平均：
#math_score       74.194221
#english_score    78.350301
#chinese_score    78.739928
```
